```python
import itertools


def effect_chart(fp, drm=0, cowering=1):
    """
    Calculate IFT effects-probabilities chart for /fp/ with /drm/ and 
    possible /cowering/ by a number of columns
    """
    def align(iterable, steps, fill):
        """
        Align /iterable/ with unsigned /steps/ leading elements filled with /fill/
        """
        return itertools.chain(
                        itertools.repeat(
                            fill, 
                            times=abs(steps)), 
                        iterable)

    # Optimized IFT
    ift = (
        ((1,  '1KIA'), (2,  '2KIA'), (6,  '3KIA'), (16, '4KIA'), (24, '5KIA'), (30, '6KIA'), (36, '7KIA')),
        ((1,  'K/1'),  (2,  '1KIA'), (6,  '2KIA'), (16, '3KIA'), (24, '4KIA'), (30, '5KIA'), (36, '6KIA')),
        ((1,  '1MC'),  (2,  'K/1'),  (4,  'K/2'),  (6,  '1KIA'), (16, '2KIA'), (24, '3KIA'), (30, '4KIA'), (36, '5KIA')),
        ((1,  '1MC'),  (4,  '2MC'),  (6,  'K/2'),  (12, 'K/3'),  (16, '1KIA'), (24, '2KIA'), (30, '3KIA'), (36, '4KIA')),
        ((1,  'NMC'),  (2,  '1MC'),  (6,  '2MC'),  (12, '3MC'),  (16, 'K/3'),  (20, 'K/4'),  (24, '1KIA'), (30, '2KIA'), (36, '3KIA')),
        ((1,  'PTC'),  (2,  'NMC'),  (4,  '1MC'),  (8,  '2MC'),  (12, '2MC'),  (16, '3MC'),  (20, '4MC'),  (24, 'K/4'),  (30, '1KIA'), (36, '2KIA')),
        ((2,  'PTC'),  (4,  'NMC'),  (6,  '1MC'),  (12, '2MC'),  (20, '3MC'),  (24, '4MC'),  (30, 'K/4'),  (36, '1KIA')),
        ((4,  'PTC'),  (6,  'NMC'),  (8,  '1MC'),  (16, '2MC'),  (24, '3MC'),  (30, '4MC'),  (36, 'K/4')),
        ((6,  'PTC'),  (8,  'NMC'),  (12, '1MC'),  (20, '2MC'),  (30, '3MC'),  (36, '4MC')),
        ((8,  'PTC'),  (12, 'NMC'),  (16, '1MC'),  (24, '2MC'),  (36, '3MC')),
        ((12, 'PTC'),  (16, 'NMC'),  (20, '1MC'),  (30, '2MC')),
        ((16, 'PTC'),  (20, 'NMC'),  (24, '1MC'),  (36, '2MC')),
        ((20, 'PTC'),  (24, 'NMC'),  (30, '1MC')),
        ((24, 'PTC'),  (30, 'NMC'),  (36, '1MC')),
        ((30, 'PTC'),  (36, 'NMC')),
        ((36, 'PTC'),)
    )

    fp_columns = (1, 2, 4, 6, 8, 12, 16, 24, 30, 36)
    cower_column = fp_columns.index(fp) - cowering
    fp_cower = fp_columns[cower_column] if cower_column >= 0 else 0
    sides = 6
    ndice = 2
    nvariants = float(sides ** ndice)
    die = xrange(1, sides + 1)
    
    # Modified dice total-probability chart
    dice_chart = [(total + drm, len(tuple(variants)) / nvariants ) 
                    for total, variants in 
                        itertools.groupby(
                            sorted(
                                itertools.product(die, repeat=ndice), key=sum), 
                            sum)]
    min_dr = dice_chart[0][0]
    
    # Non-aligned probabilities iterable
    probabilities = (row[1] for row in dice_chart)
    
    # Non-aligned effects iterable
    effects = [tuple(
                itertools.takewhile(
                    lambda cell: cell[0] <= fp, 
                    itertools.chain(
                        ((None, None),), 
                        row
                    )))[-1][1]
                             for row in ift]
    min_effect = effects[0]
    
    # Align either effects/probabilities iterable
    if min_dr > 0:
        probabilities = align(probabilities, min_dr, 0)
    elif min_dr < 0:
        effects = align(effects, min_dr, min_effect)
        
    # Effect probability chart
    return ((row[0], sum((group_row[1] for group_row in row[1]))) 
                for row in itertools.groupby(
                    itertools.ifilter(
                        lambda x: x[1] > 0, 
                        itertools.izip_longest(effects, probabilities, fillvalue=0)), 
                    lambda x: x[0]))

#def fire(fp, drm=0):
#    return ((ift[out[0]][fp], out[1]) for out in dr(drm))

print tuple(effect_chart(1, -2))
```
